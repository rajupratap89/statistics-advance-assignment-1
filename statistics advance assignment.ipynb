```python
Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_size=1000
probability_sucess=0.4
number_of_trials=10

random_sample=np.random.binomial(number_of_trials ,probability_sucess ,sample_size)
plt.hist(random_sample , bins=np.arange(number_of_trials + 2) - 0.5, edgecolor='black')
plt.xlabel('number of sucess')
plt.ylabel('frequency')
plt.title('Histogram')
plt.xticks(np.arange(number_of_trials + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
